```python
import json

js = "C:/Users/25031/Desktop/新建文件夹/data_clean/publico/news_publico.json"
with open(js, 'r', encoding='utf-8') as f:
    data = json.load(f)

dct = []
for sub in data :
    document_creation_time = sub["date"][0:10]
    dct.append(document_creation_time)

print(len(dct))
```

```python

folder_path = "C:/Users/25031/Desktop/SBERT_result_data2/"
import os
import torch
#os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'max_split_size_mb=10'

import gc
Vec = []
count = 0

for count in range(0,31962):
    # READ article file
    article_path = "C:/Users/25031/Desktop/dataset/" + str(count) + ".txt"
    file = open(article_path, 'r', encoding='utf-8')
    content = file.read()
    file.close()

    content = str(dct[count]) + content
    # process by BERT model
    with torch.no_grad():
      outputs = model([content])
    Vec.append(outputs)

    file_path = folder_path + str(count) + ".txt"
    with open(file_path, 'w', encoding='utf-8') as f:
      f.write(str(outputs))
    
    print(count, " COMPLETE \n")
    count = count + 1
    del outputs
    gc.collect()
    torch.cuda.empty_cache()
    

print(len(Vec))
```

```python
vec_numpy = []
for v in Vec :
    v = v.numpy()
    vec_numpy.append(v[0])
```

```python
import numpy
print(len(vec_numpy))
vec_numpy = numpy.array(vec_numpy)
print(type(vec_numpy))
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=1.5) #, affinity='cosine', linkage='average', distance_threshold=0.4)
clustering_model.fit(vec_numpy)
cluster_assignment = clustering_model.labels_
```

```python
print(cluster_assignment)
print(len(cluster_assignment))
```

```python
clustered_list = {}
for article_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_list:
        clustered_list[cluster_id] = []

    clustered_list[cluster_id].append(article_id)

file_path = "C:/Users/25031/Desktop/AGGLO_RESULT2/"
for n, cluster in clustered_list.items():
    print("Cluster ", n+1)
    print(cluster) 
    filename = "Cluster " + str(n+1)
    file = file_path + filename + ".txt"
    with open(file, 'w', encoding='utf-8') as f:
        f.write(str(cluster)) 
    
```

```python
import os
import re

folder_path = 'C:/Users/25031/Desktop/time/'


# 日期格式
date_pattern = re.compile(r'\d{4}-\d{2}-\d{2}')
# 所有文章的对应日期信息
data_lists = []

count = 0
for count in range(31962):
    path = folder_path + str(count) + ".txt"
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        data_list = [line.strip() for line in lines if date_pattern.match(line.strip())]
        data_lists.append(data_list)
```

```python
count = 0
for count in range(31962):
    data_lists[count].append(dct[count])    
    
print(data_lists[0:5])
```

```python
n = 0
file_path = "C:/Users/25031/Desktop/all_time/"
for n in range(31962):
    filename = "Time_of_article " + str(n)
    file = file_path + filename + ".txt"
    with open(file, 'w', encoding='utf-8') as f:
        for date in data_lists[n]:
            f.write(str(date)) 
            f.write("\n")
```

```python
date_all_cluster = []

for n, cluster in clustered_list.items():
    date_each_cluster = []
    for article_num in cluster:
        for date in data_lists[article_num]:
            date_each_cluster.append(date[0:10])
            
    # cluster的编号对应日期list        
    temp = []
    temp.append(n+1)
    temp.append(date_each_cluster)
    date_all_cluster.append(temp)
    
print(date_all_cluster[0])          
```
